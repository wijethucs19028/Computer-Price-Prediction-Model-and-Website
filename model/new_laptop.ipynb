```python
def set_processor(name):
    if name=='Intel Core i7' or name=='Intel Core i5' or name=='Intel Core i3' or name=='Intel Celeron Dual':
        return name
    else:
        if name.split()[0]=='AMD':
            return 'AMD'
        else:
            return 'Other'
data['cpu_name']=data['cpu_name'].apply(set_processor)            
```

```python
# Define the mapping of original values to their corresponding categories
mapping = {'Windows 10': 'Windows', 'Windows 7': 'Windows', 'Windows 10 S': 'Windows',
           'macOS': 'Mac', 'Mac OS X': 'Mac', 'Linux': 'Linux','Chrome OS':'Chrome'}

# Replace the original values in the 'OpSys' column with their corresponding categories
data['OpSys'] = data['OpSys'].replace(mapping)

# For the remaining values (any other values), replace them with 'Other'
data.loc[~data['OpSys'].isin(['Windows', 'Mac', 'Linux','Chrome']), 'OpSys'] = 'Other'
```

```python
x=data.drop('Price_euros',axis=1)
y=data['Price_euros']
```

```python
from sklearn.model_selection import train_test_split # type: ignore
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.24,random_state = 4)
```

```python
x_train.shape,x_test.shape
```

```python
def model_acc(model):
    model.fit(x_train,y_train)
    acc=model.score(x_test,y_test)
    print(str(model)+'-->'+str(acc))
```

```python
from sklearn.linear_model import LinearRegression # type: ignore
lr=LinearRegression()
model_acc(lr)

from sklearn.linear_model import Lasso # type: ignore
lasso=Lasso()
model_acc(lasso)

from sklearn.tree import DecisionTreeRegressor # type: ignore
dt=DecisionTreeRegressor()
model_acc(dt)

from sklearn.ensemble import RandomForestRegressor # type: ignore
rf=RandomForestRegressor()
model_acc(rf)
```

```python
#hyper parameter tunning
from sklearn.model_selection import GridSearchCV # type: ignore

parameters={'n_estimators':[10,50,100],'criterion':['squared_error','absolute_error','poisson']}
grid_obj=GridSearchCV(estimator=rf,param_grid=parameters)

grid_fit=grid_obj.fit(x_train,y_train)

new_best_model=grid_fit.best_estimator_
new_best_model



```

```python
new_best_model.score(x_test,y_test)
```

```python
# with RandomForest default parameters--->0.8268, after hyper parameter tunning---->0.8296
#now we have a model. then we have to create a website using this model.(for that we have to save this model (we can use pickle for that).)

# Save the model to a pickle file
import pickle


with open('predictor.pickle', 'wb') as f:
    pickle.dump(new_best_model, f)

print("Model saved as predictor.pickle")
```
